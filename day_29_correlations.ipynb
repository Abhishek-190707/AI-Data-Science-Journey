```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load Data
df = sns.load_dataset('diamonds')

# Inspect
print(df.head())

# 2. Calculate Correlation Matrix
# This function only works on numerical columns.
# It compares every column against every other column.
correlation = df.corr(numeric_only=True)

print(correlation['price'].sort_values(ascending=False))
# (Look at the output: Which value is closest to 1.0?)

# 3. Visualizing the Relationship (The Heatmap)
sns.heatmap(data=correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# --- STEP 4: The Proof (Regression Plot) ---
plt.figure(figsize=(10, 6))

# sns.regplot
# x = The Independent Variable (Cause) -> Carat
# y = The Dependent Variable (Effect) -> Price
# scatter_kws={'alpha':0.1} -> Makes dots transparent so you can see density
# line_kws={'color':'red'} -> Makes the trend line red for contrast

sns.regplot(x='carat', y='price', data=df,
            scatter_kws={'alpha':0.1},
            line_kws={'color':'red'})

plt.title('The Price of a Diamond: Carat vs Price')
plt.xlabel('Carat (Weight)')
plt.ylabel('Price ($)')
plt.show()
```
