```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. CREATE DATA
data = {
    "Student": ["Amit", "Neha", "Rahul", "Priya", "Ankit", "Suresh"],
    "Study_Hours": [2, 8, 3, 9, 5, 1],
    "Marks": [60, 95, 75, 98, 85, 55]
}
df = pd.DataFrame(data)

# 2. SETUP THEME (Make it look professional)
sns.set_theme(style="whitegrid")

# 3. PLOT A: SCATTER PLOT (Relationship)
# Is there a correlation between Hours and Marks?
plt.figure(figsize=(10, 5)) # Set the canvas size (10 inches wide, 5 tall)
sns.scatterplot(data=df, x="Study_Hours", y="Marks", s=200, color="red") # s = dot size

# Customizing using Matplotlib
plt.title("Study Hours vs. Marks Obtained", fontsize=16)
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.show()

# 4. PLOT B: BAR CHART (Comparison)
# Who scored the highest?
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Student", y="Marks", palette="viridis")

plt.title("Class Performance", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. PANDAS: Create the Data
data = {
    "Over": [1, 2, 3, 4, 5, 6],
    "Runs": [4, 12, 5, 8, 20, 10]
}
df = pd.DataFrame(data)

# 2. SEABORN: The "Pretty" Line Chart
# sns.set_theme() automatically adds gridlines and nice fonts
sns.set_theme(style="darkgrid")

plt.figure(figsize=(10, 5))

# We use 'lineplot' from Seaborn. It reads directly from the DataFrame.
# marker='o' adds the dots at each data point.
sns.lineplot(data=df, x="Over", y="Runs", marker='o', linewidth=2.5, color='blue')

# 3. MATPLOTLIB: The Technical Details
# We still use Matplotlib (plt) to add specific annotations
plt.title("Match Analysis: Run Flow per Over", fontsize=16)
plt.axhline(y=10, color='green', linestyle='--', label="Target Rate (10 runs)")

# Add a legend so we know what the green line is
plt.legend()

plt.show()
```
