# Day 47: Regression Inference – Foundation

## Topics Covered:
- Manual derivation of slope (covariance / variance formula)
- Computing intercept
- Generating predictions
- Residual calculation
- Residual variance estimation
- Understanding degrees of freedom (n - 2)

## Key Insights:
- Regression slope is based on covariance structure.
- Residuals measure unexplained variation.
- Residual variance estimates noise in the model.
- Dividing by (n - 2) accounts for estimating two parameters (intercept and slope).

This forms the foundation for computing standard error and statistical significance of regression coefficients.

```python
import numpy as np

# 1. Let's set up some sample data arrays
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 6])

# 2. Calculate the means (x_bar and y_bar)
x_mean = np.mean(x)
y_mean = np.mean(y)

# 3. Calculate the deviations from the mean
x_deviations = x - x_mean
y_deviations = y - y_mean

# 4. Calculate the numerator: sum of (x_i - x_mean) * (y_i - y_mean)
# NumPy multiplies the arrays element by element, then we sum them up
numerator = np.sum(x_deviations * y_deviations)

# 5. Calculate the denominator: sum of (x_i - x_mean)^2
denominator = np.sum(x_deviations**2)

# 6. Calculate the final slope (m)
m = numerator / denominator

print(f"The calculated slope (m) is: {m}")
```

```python
# np.cov returns a covariance matrix, we want the specific covariance of x and y at index [0][1]
covariance = np.cov(x, y)[0][1]

# np.var calculates the variance. ddof=1 is used for sample variance to match np.cov
variance_x = np.var(x, ddof=1)

m_alt = covariance / variance_x
print(f"The slope using cov/var is: {m_alt}")
```

```python
# Calculate the y-intercept (b)
b = y_mean - (m * x_mean)

print(f"The calculated y-intercept (b) is: {b}")

# Now you have your full equation!
print(f"The full equation for the line of best fit is: y = {m}x + {b}")
```

```python
# 1. First, calculate all the predicted y values (y-hat) using your line equation
# NumPy automatically applies this to the entire array of x values at once
y_pred = (m * x) + b

# 2. Subtract the predictions from the actual observed y values
residuals = y - y_pred

print("Actual y values:    ", y)
print("Predicted y values: ", y_pred)
print("Residuals (errors): ", residuals)
```

```python
import numpy as np

# Assuming 'residuals' is already calculated as: residuals = y - y_pred

# Square every residual in the array, then sum them all together
rss = np.sum(residuals**2)

print(f"The Residual Sum of Squares (RSS) is: {rss}")
```

```python
import numpy as np

# Assuming 'x' is your array of independent variables and 'rss' is already calculated
n = len(x)

# Calculate Residual Variance (unbiased)
residual_variance = rss / (n - 2)

print(f"The Residual Variance is: {residual_variance}")
```
