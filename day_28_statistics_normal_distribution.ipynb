```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- PART 1: The Outlier Trap (Synthetic Data) ---
# Scenario: 5 Employees
salaries = [40000, 42000, 45000, 48000, 50000]

print(f"Normal Team - Mean: {np.mean(salaries)}")
print(f"Normal Team - Median: {np.median(salaries)}")

# Scenario: The CEO joins
salaries_with_ceo = salaries + [5000000] # Adding 5 Million

print("\n--- The CEO Effect ---")
print(f"With CEO - Mean: {np.mean(salaries_with_ceo):.2f} (Huge Jump!)")
print(f"With CEO - Median: {np.median(salaries_with_ceo)} (Stable - Better Measure)")


# --- PART 2: Real World (Titanic Fares) ---
df = sns.load_dataset('titanic')

mean_fare = df['fare'].mean()
median_fare = df['fare'].median()
std_fare = df['fare'].std()

print(f"\n--- Titanic Fare Analysis ---")
print(f"Mean Fare: ${mean_fare:.2f}")
print(f"Median Fare: ${median_fare:.2f}")
print(f"Standard Deviation: ${std_fare:.2f}")

# Why the difference? Let's visualize the "Skew"
plt.figure(figsize=(10, 5))
sns.histplot(df['fare'], kde=True)
plt.axvline(mean_fare, color='red', linestyle='--', label='Mean')
plt.axvline(median_fare, color='green', linestyle='-', label='Median')
plt.title('Fare Distribution: Mean vs Median')
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Load the new dataset
df = sns.load_dataset('tips')

# Inspect it
print(df.head())

# 2. Calculate Mean and Standard Deviation
mean_tip = df['tip'].mean()
std_tip = df['tip'].std()

print(f"Average Tip: ${mean_tip:.2f}")
print(f"Standard Deviation: ${std_tip:.2f}")

# 3. Define the Threshold (3 Sigma Rule)
# Any tip higher than (Mean + 3 * Sigma) is an anomaly
upper_bound = mean_tip + ( 3 * std_tip)

print(f"Anomaly Threshold: ${upper_bound:.2f}")

# 4. Filter the Outliers
outliers = df[ df['tip'] > upper_bound]

print("\n--- The Super Tippers (Anomalies) ---")
print(outliers)

# 5. Visualization
sns.histplot(df['tip'],kde=True )
plt.axvline(upper_bound, color='red', linestyle='--')
plt.show()
```
