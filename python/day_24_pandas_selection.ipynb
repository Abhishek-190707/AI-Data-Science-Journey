```python
import pandas as pd
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 78],
    'Class': ['A', 'A', 'B']
})

# 1. iloc: Get the first row (Alice)
print(df.iloc[0])

# 2. loc: Get the Score of row 1
print(df.loc[1, 'Score'])

# 3. Filtering: Get all students in Class 'A' with Score > 80
top_students = df[ (df['Class'] == 'A') & (df['Score'] > 80) ]
print(top_students)
```

```python
import pandas as pd

# 1. Load Data directly from GitHub using Pandas
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Check the column names first so you know what to type in .loc
print(df.columns)
```

```python
manifest = df.loc[:,['Survived','Pclass','Sex','Age']]
print(manifest)
```

```python
subset = df.iloc[0:10 , 1:5]
print(subset)
```

```python
rich_passengers = pd.DataFrame(df[df['Fare'] > 100])
print(rich_passengers)
```

```python
priority_rescue = df[(df['Pclass'] == 1) & (df['Sex'] == 'female')]
print("Priority Rescue Count:", len(priority_rescue))
```
