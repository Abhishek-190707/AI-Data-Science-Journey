## Worksheet Overview

This notebook demonstrates linear regression using the Normal Equation. It covers the following steps:

*   **Data Initialization**: Setting up the feature matrix (X) and target vector (y).
*   **Weight Calculation**: Computing the optimal weights (w) using the Normal Equation.
*   **Prediction Generation**: Making predictions based on the calculated weights.
*   **Linear Regression Function**: Encapsulating the Normal Equation logic into a reusable function.

```python
import numpy as np

# Feature matrix X, including a bias term (column of ones) and a single feature column.
# The first column `[1, 1, 1, 1]` represents the bias term.
# The second column `[1, 2, 3, 4]` represents the actual feature values.
X = np.array([[1, 1],
              [1, 2],
              [1, 3],
              [1, 4]])

# Target vector y, representing the corresponding output values for each set of features.
y = np.array([2, 3, 5, 4])
```

```python
X_T = X.T # Transpose of the feature matrix X, required for the Normal Equation.

# Calculate the optimal weights (w) using the Normal Equation.
# The Normal Equation is: w = (X^T * X)^-1 * X^T * y
# Where:
# - X_T is the transpose of the feature matrix X.
# - np.linalg.inv() calculates the inverse of a matrix.
# - @ performs matrix multiplication.
w = np.linalg.inv(X_T @ X) @ X_T @ y

# 'w' represents the optimal weights (coefficients) of the linear regression model.
# The first element of 'w' is the bias term (intercept), and subsequent elements are coefficients for each feature.
print("Weights:", w)
```

```python
predictions = X @ w # Calculate predictions by multiplying the feature matrix X with the learned weights w.
# Each element in 'predictions' is the predicted target value for the corresponding row in X.
print("Predictions:", predictions)
```

```python
def linear_regression(X, y):
    """
    Computes the optimal weights for a linear regression model using the Normal Equation.

    Parameters:
    X (numpy.ndarray): The feature matrix, including the bias term (a column of ones).
                       Each row represents a data point, and each column represents a feature.
    y (numpy.ndarray): The target vector, containing the true output values for each data point.

    Returns:
    numpy.ndarray: A vector of optimal weights (coefficients), where the first element is the
                   bias (intercept) and subsequent elements are the coefficients for each feature.
    """
    # Step 1: Transpose the feature matrix X. X_T is used in the Normal Equation.
    X_T = X.T

    # Step 2: Calculate the optimal weights (w) using the Normal Equation:
    # w = (X^T * X)^-1 * X^T * y
    # - (X_T @ X): Matrix multiplication of X transpose and X.
    # - np.linalg.inv(...): Calculates the inverse of the resulting matrix.
    # - (... @ X_T): Matrix multiplication of the inverse with X transpose.
    # - (... @ y): Final matrix multiplication with the target vector y.
    w = np.linalg.inv(X_T @ X) @ X_T @ y

    return w

    ### Insight

'''Linear regression is solving a system of equations in matrix form.
The normal equation uses matrix multiplication and inversion to compute optimal weights.'''
```

```python
import matplotlib.pyplot as plt

# Call the linear_regression function to compute the weights
w_calculated = linear_regression(X, y)

# Calculate predictions using the calculated weights
predictions_calculated = X @ w_calculated

# Create a scatter plot of the original data points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 1], y, label='Original Data', color='blue')

# Plot the linear regression line
plt.plot(X[:, 1], predictions_calculated, label='Linear Regression Fit', color='red')

# Add plot title and labels
plt.title('Linear Regression Fit')
plt.xlabel('Feature Value')
plt.ylabel('Target Value')

# Add a legend to distinguish between original data and the regression line
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
