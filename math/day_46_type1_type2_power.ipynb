## Understanding Statistical Power: A Simulation Study

This worksheet explores the concept of statistical power through a series of simulations. Statistical power is the probability that a hypothesis test correctly rejects the null hypothesis when the alternative hypothesis is true. In simpler terms, it's the ability of a test to detect an effect if an effect truly exists.

### Key Concepts Illustrated:

*   **Original Scenario (Cell 1):** We start with a baseline simulation to estimate statistical power given a set of initial parameters: sample size (`n=30`), true mean (`true_mean=72`), and significance level (`alpha=0.05`). This provides a reference point for comparison.

*   **Impact of Sample Size (Cell 2):** This simulation demonstrates how increasing the sample size (`n`) significantly improves the power of the test. A larger sample provides more information, making it easier to detect a true effect, thus reducing the chance of a Type II error (failing to reject a false null hypothesis).

*   **Impact of Effect Size (Cell 3):** Here, we observe the effect of a larger difference between the true mean (`true_mean`) and the population mean (`population_mean`). A greater effect size makes it easier for the test to distinguish between the null and alternative hypotheses, leading to higher statistical power.

*   **Impact of Significance Level (Alpha) (Cell 4):** This simulation illustrates the trade-off associated with the significance level (`alpha`). Reducing `alpha` (making the test more stringent) decreases the probability of a Type I error (falsely rejecting the null hypothesis). However, this often comes at the cost of reduced statistical power, increasing the risk of a Type II error.

**In summary, these simulations collectively highlight:**
*   **Higher sample size leads to higher power.**
*   **Larger effect size leads to higher power.**
*   **Lower alpha (more conservative test) leads to lower power.**

Understanding these relationships is crucial for designing effective experiments and interpreting the results of hypothesis tests.

```python
import numpy as np
import scipy.stats as stats

np.random.seed(42)

population_mean = 70
true_mean = 72
population_std = 10
n = 30
alpha = 0.05

reject_count = 0
trials = 1000

for _ in range(trials):
    sample = np.random.normal(true_mean, population_std, n)
    t_stat, p_value = stats.ttest_1samp(sample, population_mean)

    if p_value < alpha:
        reject_count += 1

power = reject_count / trials

print("Estimated Power:", power)
# This is the original code cell.
# Parameters: n=30, true_mean=72, alpha=0.05
# Estimated Power: 0.184
```

```python
import numpy as np
import scipy.stats as stats

np.random.seed(12)

population_mean = 70
true_mean = 72
population_std = 10
n = 100  # Sample size increased from 30 to 100
alpha = 0.05

reject_count = 0
trials = 1000

for _ in range(trials):
    sample = np.random.normal(true_mean, population_std, n)
    t_stat, p_value = stats.ttest_1samp(sample, population_mean)

    if p_value < alpha:
        reject_count += 1

power = reject_count / trials

print("Estimated Power:", power)
# This cell shows the effect of increasing the sample size (n).
# As 'n' increased to 100 (from 30), the estimated power increased to 0.512 (from 0.184).
```
