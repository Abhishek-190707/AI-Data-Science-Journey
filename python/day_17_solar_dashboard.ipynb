```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate Data
time = np.linspace(0, 10, 100)
velocity = time**2                      # Exponential speed
vibration = np.random.normal(5, 1, 100) # Noise/vibration data
fuel_types = ['Electric', 'Petrol', 'Diesel']
efficiency = [90, 35, 45]

# 2. Setup the Dashboard
plt.figure(figsize=(10, 8)) # Make the total window larger
plt.style.use('ggplot')

# --- Plot 1: Velocity Trend (Top) ---
plt.subplot(2, 1, 1) # 2 rows, 1 col, Position 1
plt.plot(time, velocity, color='blue', linewidth=2)
plt.title("Engine Velocity Over Time")
plt.ylabel("km/h")

# --- Plot 2: Vibration Distribution (Bottom Left) ---
plt.subplot(2, 2, 3) # 2 rows, 2 cols, Position 3
plt.hist(vibration, bins=15, color='orange', edgecolor='black')
plt.title("Engine Vibration Stability")
plt.xlabel("G-Force")

# --- Plot 3: Fuel Efficiency (Bottom Right) ---
plt.subplot(2, 2, 4) # 2 rows, 2 cols, Position 4
plt.bar(fuel_types, efficiency, color=['green', 'red', 'gray'])
plt.title("Efficiency Comparison (%)")

# 3. Final Polish
plt.tight_layout() # IMPORTANT: Prevents titles from overlapping
plt.savefig('engine_dashboard.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Setup Theme and Figure
plt.style.use('seaborn-v0_8-darkgrid') # Gives it a clean, modern look
plt.figure(figsize=(12, 10)) # Increased size for better visual spacing

# --- Plot 1: Energy Generation (Line Chart) ---
plt.subplot(2, 1, 1)
# Adding markers and a dashed line for a "technical" feel
plt.plot([6, 12, 18, 24], [100, 200, 330, 450],
         color='#FFD700', linewidth=3, marker='o', markersize=8, label='Energy Output')
plt.fill_between([6, 12, 18, 24], [100, 200, 330, 450], color='#FFD700', alpha=0.2) # Shaded area
plt.xlabel("Time (24-Hour Format)", fontweight='bold')
plt.ylabel("Energy (kWh)", fontweight='bold')
plt.title("Solar Energy Generation Trend", fontsize=14, pad=10)
plt.grid(True, linestyle='--', alpha=0.6)

# --- Plot 2: Voltage Fluctuations (Histogram) ---
plt.subplot(2, 2, 3)
n, bins, patches = plt.hist(np.random.normal(15, 1, 1000), bins=20,
                            color='#1E90FF', edgecolor='white', alpha=0.8)
plt.title("Grid Voltage Stability", fontsize=12)
plt.xlabel("Voltage (V)")
plt.ylabel("Frequency")

# --- Plot 3: Battery Storage (Bar Chart) ---
plt.subplot(2, 2, 4)
bars = plt.bar(["Bat A", "Bat B", "Bat C", "Bat D"], [70, 100, 90, 45],
               color=['#4CAF50', '#4CAF50', '#FF9800', '#F44336']) # Color code by health
plt.title("Battery Storage Levels (%)", fontsize=12)
plt.xlabel("Storage Unit")
plt.ylabel("Capacity")
plt.ylim(0, 110) # Set limit so bars don't hit the top

# 2. The Professional Touch
plt.suptitle("SOLAR POWER STATION: SYSTEM DIAGNOSTICS", fontsize=18, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make room for suptitle

# 3. Export
plt.savefig('solar_dashboard_v2.png', dpi=300) # dpi=300 makes it high resolution
plt.show()
```
