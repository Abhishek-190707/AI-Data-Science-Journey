## Summary of Simulations and Visualizations

This notebook performed several simulations to demonstrate basic probability and statistics concepts using `numpy` and `matplotlib`.

1.  **Dice Roll Simulation**: We simulated 10,000 rolls of a 6-sided die. The mean and variance were calculated, and the distribution was visualized using a histogram. As expected, the histogram showed a relatively uniform distribution across the 6 faces, with the mean approaching 3.5.

2.  **Coin Flip Simulation**: A coin was flipped 10,000 times with a specified probability of 70% for heads (represented by 1). The mean of the results (proportion of heads) was calculated, which closely approximated the set probability of 0.7, illustrating the Law of Large Numbers.

3.  **Exam Marks Simulation**: We simulated 10,000 exam marks following a normal distribution with a mean of 70 and a standard deviation of 10. The mean and variance of the simulated marks were calculated, and a histogram was used to visualize the bell-shaped curve characteristic of a normal distribution. This demonstrated how to model real-world data like exam scores.

```python
# This cell simulates 10,000 rolls of a 6-sided die and calculates its basic statistics.
import numpy as np

# np.random.randint(low, high, size) simulates 'size' number of rolls, generating
# integers from 'low' (inclusive) to 'high' (exclusive). Here, it simulates
# 10,000 rolls of a die with faces 1 through 6.
dice = np.random.randint(1, 7, 10000)

# np.mean() calculates the average value of the simulated dice rolls.
print("Mean:", np.mean(dice))
# np.var() calculates the variance of the simulated dice rolls, which measures
# the spread or dispersion of the data points around the mean.
print("Variance:", np.var(dice))
```

```python
# This cell simulates 10,000 coin flips with a specified probability for heads (0.7).
import numpy as np

# np.random.choice(a, size, p) simulates 'size' number of choices from array 'a',
# with 'p' being the probabilities of each choice. Here, it simulates 10,000 coin flips,
# where 0 represents tails and 1 represents heads. The probability of getting heads (1) is 0.7,
# and tails (0) is 0.3.
coin = np.random.choice([0,1], size=10000, p=[0.3,0.7])

# np.mean(coin) calculates the proportion of '1's (heads) in the simulation,
# which should approximate the given probability of 0.7 due to the Law of Large Numbers.
print("Mean:", np.mean(coin))
```

```python
# This cell visualizes the distribution of the simulated dice rolls.
import matplotlib.pyplot as plt # Import the matplotlib.pyplot library for plotting, commonly aliased as plt.

plt.hist(dice, bins=6) # plt.hist() creates a histogram of the 'dice' data, with 'bins=6' for the 6 possible outcomes.
plt.title("Dice Distribution") # plt.title() sets the title of the plot.
plt.show() # plt.show() displays the generated plot.
```

```python
# This cell simulates 10,000 exam marks following a normal distribution and visualizes them.
import numpy as np
import matplotlib.pyplot as plt

# np.random.normal(loc, scale, size) generates 'size' number of samples from a normal distribution.
# 'loc' is the mean (average score) and 'scale' is the standard deviation (spread of scores).
# Here, it simulates 10,000 exam marks with a mean of 70 and a standard deviation of 10.
marks = np.random.normal(loc=70, scale=10, size=10000)

# np.mean() calculates the average of the simulated exam marks.
print("Mean:", np.mean(marks))
# np.var() calculates the variance of the simulated exam marks, indicating the spread.
print("Variance:", np.var(marks))

# plt.hist() creates a histogram of the 'marks' data, with 'bins=30' to show the distribution shape.
plt.hist(marks, bins=30)
# plt.title() sets the title of the plot.
plt.title("Simulated Exam Marks Distribution")
# plt.show() displays the generated plot.
plt.show()
```
