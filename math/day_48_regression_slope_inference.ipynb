## Topics Covered:

- t-statistic for testing β₁ = 0

## Key Learnings:
- Standard Error measures uncertainty in slope estimate.
- t-statistic represents signal-to-noise ratio.
- p-value indicates strength of evidence against H₀.
- Confidence interval shows plausible range of slope values.
- If CI excludes 0, the feature is statistically significant.

## Insight:
Regression is not just fitting a line.
It is estimating a relationship with quantified uncertainty.

```python
import numpy as np

# 1. Let's set up some sample data arrays
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 6])

# 2. Calculate the means (x_bar and y_bar)
x_mean = np.mean(x)
y_mean = np.mean(y)

# 3. Calculate the deviations from the mean
x_deviations = x - x_mean
y_deviations = y - y_mean

# 4. Calculate the numerator: sum of (x_i - x_mean) * (y_i - y_mean)
# NumPy multiplies the arrays element by element, then we sum them up
numerator = np.sum(x_deviations * y_deviations)

# 5. Calculate the denominator: sum of (x_i - x_mean)^2
denominator = np.sum(x_deviations**2)

# 6. Calculate the final slope (m)
m = numerator / denominator

print(f"The calculated slope (m) is: {m}")
```

```python
# np.cov returns a covariance matrix, we want the specific covariance of x and y at index [0][1]
covariance = np.cov(x, y)[0][1]

# np.var calculates the variance. ddof=1 is used for sample variance to match np.cov
variance_x = np.var(x, ddof=1)

m_alt = covariance / variance_x
print(f"The slope using cov/var is: {m_alt}")
```

```python
# Calculate the y-intercept (b)
b = y_mean - (m * x_mean)

print(f"The calculated y-intercept (b) is: {b}")

# Now you have your full equation!
print(f"The full equation for the line of best fit is: y = {m}x + {b}")
```

```python
# 1. First, calculate all the predicted y values (y-hat) using your line equation
# NumPy automatically applies this to the entire array of x values at once
y_pred = (m * x) + b

# 2. Subtract the predictions from the actual observed y values
residuals = y - y_pred

print("Actual y values:    ", y)
print("Predicted y values: ", y_pred)
print("Residuals (errors): ", residuals)
```

```python
import numpy as np

# Assuming 'residuals' is already calculated as: residuals = y - y_pred

# Square every residual in the array, then sum them all together
rss = np.sum(residuals**2)

print(f"The Residual Sum of Squares (RSS) is: {rss}")
```

```python
import numpy as np

# Assuming 'x' is your array of independent variables and 'rss' is already calculated
n = len(x)

# Calculate Residual Variance (unbiased)
residual_variance = rss / (n - 2)

print(f"The Residual Variance is: {residual_variance}")
```

```python
# Calculate the Standard Error of the Slope
# We simply divide the residual variance by our original 'denominator'
# and take the square root of the whole thing.
se_slope = np.sqrt(residual_variance / denominator)

print(f"The Standard Error of the Slope is: {se_slope}")
```

```python
# Calculate the t-statistic
t_stat = m / se_slope

print(f"The t-statistic for the slope is: {t_stat}")
```

```python
# To find the p-value for a two-tailed test:
# 1. Take the absolute value of the t-statistic (so it works for negative slopes too)
# 2. Use stats.t.sf (Survival Function, which calculates the area in one tail)
# 3. Multiply by 2 because it is a two-tailed test
import scipy.stats as stats

# 1. We must actively define 'n' and 'degrees_of_freedom'
n = len(x)
degrees_of_freedom = n - 2

p_value = stats.t.sf(np.abs(t_stat), df=degrees_of_freedom) * 2

print(f"The calculated p-value is: {p_value}")

# A simple check for significance
if p_value < 0.05:
    print("Result: The slope is statistically significant (p < 0.05).")
else:
    print("Result: The slope is NOT statistically significant (p >= 0.05).")
```

```python
# 2. Set the confidence level (95% is standard)
confidence_level = 0.95
alpha = 1.0 - confidence_level

# 3. Find the critical t-value (t*) for a two-tailed test
# We split alpha in half (0.025 per tail) so we find the point at 0.975
t_critical = stats.t.ppf(1.0 - (alpha / 2), df=degrees_of_freedom)

# 4. Calculate the Margin of Error
margin_of_error = t_critical * se_slope

# 5. Calculate the Lower and Upper bounds
lower_bound = m - margin_of_error
upper_bound = m + margin_of_error

print(f"Calculated Slope (m): {m:.4f}")
print(f"Critical t-value: {t_critical:.4f}")
print(f"Margin of Error: ±{margin_of_error:.4f}")
print(f"95% Confidence Interval: [{lower_bound:.4f}, {upper_bound:.4f}]")
```
