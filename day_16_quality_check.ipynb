```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate Data (Simulating 1000 student scores)
# We use a "Normal Distribution" (Bell Curve)
# Mean (average) = 75, Standard Deviation (spread) = 15
scores = np.random.normal(75, 15, 1000)

# 2. Setup the Plot Style
plt.style.use('fivethirtyeight')

# 3. Create the Histogram
# bins=20 means "divide the scores into 20 distinct buckets"
# edgecolor='black' adds a border so you can see separate bars
plt.hist(scores, bins=20, color='#6d904f', edgecolor='black')

# 4. Add Labels
plt.title('Distribution of Student Scores')
plt.xlabel('Score (0-100)')
plt.ylabel('Number of Students')

# 5. Add a vertical line for the Passing Grade (40)
plt.axvline(40, color='red', label='Passing Line', linewidth=2)
plt.legend()

# 6. Save & Show
plt.tight_layout()
plt.savefig('scores_histogram.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Generate Data (Simulating Factory Production)
# We produce 500 bolts.
# Target diameter = 10.0 mm (Mean)
# Machine Error = 0.1 mm (Standard Deviation)
bolts = np.random.normal(10.0, 0.1, 500)

# 2. Setup the Visual Style
plt.style.use('fivethirtyeight')

# 3. Create the Histogram
# bins=15: Divide the 500 bolts into 15 size categories
plt.hist(bolts, bins=15, color='#5DADE2', edgecolor='black', label='Produced Bolts')

# 4. Add Quality Control Lines (Tolerance Limits)
# Any bolt outside these red lines is "Defective"
plt.axvline(9.8, color='red', linestyle='--', linewidth=2, label='Min Tolerance (9.8)')
plt.axvline(10.2, color='red', linestyle='--', linewidth=2, label='Max Tolerance (10.2)')

# 5. Labels and Titles
plt.title('Bolt Production Quality')
plt.xlabel('Diameter (mm)')
plt.ylabel('Count of Bolts')
plt.legend(loc='upper right')

# 6. Save and Show
plt.tight_layout()
plt.savefig('bolt_quality.png')
plt.show()
```
