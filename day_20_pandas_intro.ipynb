```python
import pandas as pd
import numpy as np

# 1. CREATE DATA (The "Dictionary" Method)
# This is how you manually build a DataFrame
data = {
    "Student": ["Amit", "Neha", "Rahul", "Priya", "Ankit"],
    "Math": [85, 92, 78, 95, 88],
    "Science": [90, 94, 82, 98, 85],
    "English": [88, 85, 80, 92, 87]
}

# Convert Dictionary -> DataFrame
df = pd.DataFrame(data)

# 2. INSPECT (The "Interview")
print("--- HEAD (First 5 rows) ---")
print(df.head())

print("\n--- INFO (Data Types & Missing Values) ---")
print(df.info())

print("\n--- DESCRIBE (Statistics) ---")
# This automatically calculates Mean, Min, Max, and Percentiles
print(df.describe())
```

```python
import pandas as pd

# 1. SETUP: Create the DataFrame
data = {
    "Student": ["Amit", "Neha", "Rahul", "Priya", "Ankit"],
    "Math": [85, 92, 78, 95, 88],
    "Science": [90, 94, 82, 98, 85],
    "English": [88, 85, 80, 92, 87]
}
df = pd.DataFrame(data)

# 2. FEATURE ENGINEERING (Creating New Columns)
# Notice how we add entire columns together instantly (Vectorization)
df['Total'] = df['Math'] + df['Science'] + df['English']
df['Average'] = df['Total'] / 3

# 3. FILTERING (Finding the Toppers)
# We create a new DataFrame 'toppers' containing only rows where Average > 90
toppers = df[df['Average'] > 90]

# 4. OUTPUT
print("--- Full Marksheet ---")
print(df)

print("\n--- ðŸ† Top Performers (Avg > 90) ---")
print(toppers)
```
