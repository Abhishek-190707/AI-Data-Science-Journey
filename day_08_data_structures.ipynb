```python
website_info = ("www.google.com", "8.8.8.8")
# tuple is safer then others becuase is it immutable , so the data inside it is safe and cannot be changed.
website_info[1] = "1.1.1.1"
```

```python
raw_ids = [101,102,101,103,104,102,105]
raw_ids_set = set(raw_ids)
# converting the list to set removes the duplicate items and leaves only unique ones
print(raw_ids_set)
print(len(raw_ids_set))

```

```python
shop_inventory = {"Apples" : 50 , "Bananas" : 20 , "Oranges" : 40}
# .get helps to get the value of a specific key
print(shop_inventory.get("Apples"))
```

```python
shop_inventory = {"Apples" : 50 , "Bananas" : 20 , "Oranges" : 40}
# .update helps to update the dictionary with a key and its respective value
shop_inventory.update({"mango" : 100})
print(shop_inventory)
```

```python
shop_inventory = {"Apples" : 50 , "Bananas" : 20 , "Oranges" : 40}
# dict[""] = "" , helps to change the value of an key
shop_inventory["Bananas"] = 30
print(shop_inventory)
```
