```python
with open('diary.txt' , 'w') as f:
  f.write("day 09 - i am learning to save data permanently.\n")
with open('diary.txt' , 'r') as r:
  read = r.read()
  print(read)
```

```python
with open('diary.txt' , 'a') as a:
  a.write("this line was added later as append \n")
with open('diary.txt' , 'r') as r:
    read = r.read()
    print(read)
```
