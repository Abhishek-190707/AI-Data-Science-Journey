```python
# --- STEP 1: Define the Function ---
def process_student(line):
    # 1. Remove the invisible newline character at the end (\n)
    clean_line = line.strip()

    # 2. Split the line into a list (using comma)
    parts = clean_line.split(',')  # Use .split(',')

    # 3. The first item is the Name
    name = parts[0]

    # 4. The rest of the items are Marks (Slice and Map to int)
    marks = list(map(int, parts[1:])) # Slice parts from index 1 to end

    # 5. Calculate Average
    average = round(sum(marks) / len(marks), 2)
# This rounds the actual number to 2 decimal places.

    # 6. Determine Status (Logic)
    if average >= 50:
        status = "PASS"
    else:
        status = "FAIL"

    # 7. Return the formatted string
    return f"{name}: {average} - {status}\n"

# --- STEP 2: The Main Execution ---
try:
    with open("/raw_grades.txt", "r") as input_file:
        with open("final_report.txt", "w") as output_file:

            # Loop through every line in the input file
            for line in input_file:
                # Call your function for THIS line
                result = process_student(line)

                # Write the result to the new file
                output_file.write(result)

    print("Success! Report generated.")

except FileNotFoundError:
    print("Error: The file raw_grades.txt does not exist.")
```
