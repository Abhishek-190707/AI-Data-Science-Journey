```python
def calculate_tax(price , tax_rate):
    """
    Calculates the final price of an item after applying tax.
    
    Formula: Total = Price + (Price * Tax Rate)
    
    Args:
        price (float): The base price of the item.
        tax_rate (float): The tax percentage expressed as a decimal (e.g., 0.18 for 18%).
        
    Returns:
        float: The calculated total cost.
    """
    tax_amount = price * tax_rate
    total = price + tax_amount
    return total
calculate_tax(100 , .10)
```

```python
def is_even(number):
    """
    Determines if a given integer is even or odd using the modulo operator.
    
    Args:
        number (int): The integer to check.
        
    Returns:
        bool: True if the number is even, False if it is odd.
    """
    if number % 2 == 0:
        return True
    else:
        return False
is_even(40)    
        
```

```python
num_to_check = [10,20,25,34,35,56,70]

print("--- Initiating Even Number Scan ---")

def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False

# Iterate through the dataset and apply the filter function
for number in num_to_check:
    
    # Pass the current number to our helper function
    if is_even(number):
        # If True, log the result to the console
        print(f"Match Found: {number} is an even number.")

print("--- Scan Complete ---")

    
```
