```python
# ==========================================
# 1. DEFINE THE CONVERSION LOGIC
# ==========================================

def km_to_miles(km):
    """Converts Kilometers to Miles"""
    return km * 0.621371

def miles_to_km(miles):
    """Converts Miles to Kilometers"""
    return miles / 0.621371

def celsius_to_fahrenheit(c):
    """Converts Celsius to Fahrenheit"""
    return (c * 9/5) + 32

def fahrenheit_to_celsius(f):
    """Converts Fahrenheit to Celsius"""
    return (f - 32) * 5/9

def kg_to_pounds(kg):
    """Converts Kilograms to Pounds"""
    return kg * 2.20462

```

```python
# ==========================================
# 2. DEFINE THE MENU DISPLAY
# ==========================================

def show_menu():
    print("\n--- ğŸ”„ MASTER UNIT CONVERTER ğŸ”„ ---")
    print("1. Kilometers to Miles")
    print("2. Miles to Kilometers")
    print("3. Celsius to Fahrenheit")
    print("4. Fahrenheit to Celsius")
    print("5. Kilograms to Pounds")
    print("Q. Quit Application")
    print("-----------------------------------")
```

```python
# ==========================================
# 3. MAIN PROGRAM LOOP
# ==========================================

print("Welcome to the Unit Converter!")

while True:
    # 1. Show the menu
    show_menu()

    # 2. Get user choice
    choice = input("Select an option (1-5 or Q): ").upper()

    # 3. Check the 'Exit' condition FIRST
    if choice == 'Q':
        print("\nExiting program... Goodbye! ğŸ‘‹")
        break # This stops the infinite loop

    # 4. Handle the math logic based on choice
    # Option 1: Km -> Miles
    elif choice == '1':
        val = float(input("Enter distance in Kilometers: "))
        result = km_to_miles(val)
        print(f"\nâœ… Result: {val} km = {result:.2f} miles")

    # Option 2: Miles -> Km
    elif choice == '2':
        val = float(input("Enter distance in Miles: "))
        result = miles_to_km(val)
        print(f"\nâœ… Result: {val} miles = {result:.2f} km")

    # Option 3: C -> F
    elif choice == '3':
        val = float(input("Enter temperature in Celsius: "))
        result = celsius_to_fahrenheit(val)
        print(f"\nâœ… Result: {val}Â°C = {result:.2f}Â°F")

    # Option 4: F -> C
    elif choice == '4':
        val = float(input("Enter temperature in Fahrenheit: "))
        result = fahrenheit_to_celsius(val)
        print(f"\nâœ… Result: {val}Â°F = {result:.2f}Â°C")

    # Option 5: Kg -> Lbs (Your Custom Feature!)
    elif choice == '5':
        val = float(input("Enter weight in Kilograms: "))
        result = kg_to_pounds(val)
        print(f"\nâœ… Result: {val} kg = {result:.2f} lbs")

    # 5. Error Handling (for bad inputs like 'X' or '9')
    else:
        print("\nâŒ Invalid selection. Please try again.")
```
