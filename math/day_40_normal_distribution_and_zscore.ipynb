# Notebook Overview

This notebook demonstrates basic statistical analysis, specifically focusing on generating a normal distribution, calculating Z-scores, and identifying outliers. It utilizes `numpy` for numerical operations and `matplotlib.pyplot` for data visualization.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10000 random numbers from a normal distribution
# with a mean of 70 and a standard deviation of 10.
data = np.random.normal(70, 10, 10000)

# Create a histogram of the generated data with 30 bins.
plt.hist(data, bins=30)
# Set the title of the histogram.
plt.title("Normal Distribution")
# Display the plot.
plt.show()
```

```python
x = 90
# Calculate the mean of the 'data' array.
mu = np.mean(data)
# Calculate the standard deviation of the 'data' array.
sigma = np.std(data)

# Calculate the Z-score for the value 'x'.
# The Z-score measures how many standard deviations an element is from the mean.
z = (x - mu) / sigma
print("Z-score:", z)
```

```python
# Calculate the Z-scores for all data points.
z_scores = (data - np.mean(data)) / np.std(data)

# Identify outliers as data points where the absolute Z-score is greater than 3.
# This means values are more than 3 standard deviations away from the mean.
outliers = data[np.abs(z_scores) > 3]

# Print the number of extreme values (outliers) found.
print("Number of extreme values:", len(outliers))
```

# Summary of Analysis

In this worksheet, we performed the following steps:

1.  **Generated a Normal Distribution**: We created a dataset of 10,000 random numbers following a normal distribution with a mean of 70 and a standard deviation of 10 using `np.random.normal()`.
2.  **Visualized the Distribution**: A histogram was plotted using `matplotlib.pyplot` to visually represent the generated normal distribution.
3.  **Calculated Z-score for a Specific Value**: We calculated the Z-score for a specific data point (`x = 90`) to determine how many standard deviations it is away from the mean of the generated dataset.
4.  **Identified Outliers**: We computed Z-scores for all data points and identified 'extreme values' or outliers as those with an absolute Z-score greater than 3, indicating values more than three standard deviations from the mean. The number of such outliers was then printed.
