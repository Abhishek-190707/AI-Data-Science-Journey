## Overview of the Worksheet

This worksheet serves as an interactive demonstration of fundamental concepts in probability and statistics, specifically the **Law of Large Numbers** and the **Central Limit Theorem**.

It explores these concepts through several simulations and visualizations:

1.  **Coin Toss Simulation:** Illustrates how the running mean of coin tosses converges to the theoretical probability (0.5) as the number of tosses increases.
2.  **Dice Roll Simulation:** Similar to the coin toss, this demonstrates the convergence of the running mean of dice rolls to its theoretical expected value (3.5).
3.  **Law of Large Numbers with Random Integers:** A more in-depth look at the Law of Large Numbers, showing the convergence of running means for different sample sizes drawn from a uniform distribution of integers between 1 and 9.
4.  **Original Uniform Distribution:** Visualizes a simple uniform distribution (numbers between 0 and 1) which serves as the basis for the Central Limit Theorem demonstration.
5.  **Central Limit Theorem Demonstration:** Shows that the distribution of sample means, taken from the uniform distribution, approximates a normal distribution, even if the original distribution is not normal.

Each section includes plots to visually confirm these important statistical principles.

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 10,000 coin tosses (0 for tails, 1 for heads)
coin = np.random.choice([0, 1], size=10000)

# Calculate the running mean of the coin tosses
# np.cumsum provides the cumulative sum, and np.arange creates the divisor for each step
running_mean = np.cumsum(coin) / np.arange(1, 10001)

# Plot the running mean
plt.plot(running_mean)

# Add a horizontal line at y=0.5 (the theoretical probability of heads)
plt.axhline(y=0.5, color='r', linestyle='--')

# Add title and labels for clarity
plt.title('Running Mean of Coin Toss')
plt.xlabel('Number of Tosses')
plt.ylabel('Running Mean')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate 10,000 dice rolls (integers from 1 to 6)
dice = np.random.randint(1, 7, 10000)

# Calculate the running mean of the dice rolls
# np.cumsum provides the cumulative sum, and np.arange creates the divisor for each step
running_dice_mean = np.cumsum(dice) / np.arange(1, 10001)

# Plot the running mean
plt.plot(running_dice_mean)

# Add a horizontal line at y=3.5 (the theoretical expected value of a fair die roll)
plt.axhline(y=3.5, color='r', linestyle='--')

# Add title and labels for clarity
plt.title('Running Mean of Dice Toss')
plt.xlabel('Number of Tosses')
plt.ylabel('Running Mean')

# Display the plot
plt.show()
```

```python
# AI used for styling but idea and implimentation was done by me,

import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility to ensure the results are consistent each time the code runs
np.random.seed(42)

# Generate data for three different sample sizes from a uniform distribution (integers 1 to 9 inclusive)
# The theoretical mean for this distribution is (1+9)/2 = 5.0
data_1 = np.random.randint(1, 10, 50)    # Small sample size
data_2 = np.random.randint(1, 10, 500)   # Medium sample size
data_3 = np.random.randint(1, 10, 5000)  # Large sample size

# Calculate the running means for each dataset
# np.cumsum computes the cumulative sum, and np.arange creates the sequence of divisors (1 to N)
running_data_1_mean = np.cumsum(data_1) / np.arange(1, 51)
running_data_2_mean = np.cumsum(data_2) / np.arange(1, 501)
running_data_3_mean = np.cumsum(data_3) / np.arange(1, 5001)

# Configure the plot for better visualization
plt.figure(figsize=(12, 6))
plt.xscale('log') # Use a log scale for the x-axis to better show convergence over a wide range of sample sizes

# Plot the running means for the three sample sets
plt.plot(np.arange(1, 51), running_data_1_mean, label='Sample Size: 50', color='#1f77b4', alpha=0.9, linewidth=2)
plt.plot(np.arange(1, 501), running_data_2_mean, label='Sample Size: 500', color='#ff7f0e', alpha=0.8, linewidth=1.5)
plt.plot(np.arange(1, 5001), running_data_3_mean, label='Sample Size: 5000', color='#2ca02c', alpha=0.7, linewidth=1)

# Add a horizontal line representing the theoretical mean (5.0)
plt.axhline(y=5.0, color='red', linestyle='--', linewidth=2, label='Theoretical Mean (5.0)')

# Apply styling for a professional look
plt.title('Law of Large Numbers: Convergence to Mean (y=5.0)', fontsize=14, fontweight='bold')
plt.xlabel('Number of Samples (N) - Log Scale', fontsize=12)
plt.ylabel('Running Mean Value', fontsize=12)
plt.ylim(1, 9) # Set y-axis limits to match the data range (1 to 9)
plt.grid(True, which="both", ls="-", alpha=0.3) # Add a grid for easier reading of values
plt.legend(loc='upper right', frameon=True, shadow=True) # Display legend to identify plots

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show() # Display the plot
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 10,000 random numbers from a uniform distribution between 0 and 1
data = np.random.uniform(0, 1, 10000)

# Plot a histogram of the original uniform distribution
# This shows that all values within the range have roughly equal probability
plt.hist(data, bins=30) # Using 30 bins to show the distribution shape
plt.title("Original Distribution (Uniform)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize an empty list to store the means of samples
sample_means = []

# Perform a simulation to demonstrate the Central Limit Theorem
# Repeat 1000 times:
for _ in range(1000):
    # 1. Take a random sample of size 30 from the 'data' (original uniform distribution)
    sample = np.random.choice(data, 30)
    # 2. Calculate the mean of this sample
    sample_means.append(np.mean(sample))

# Plot a histogram of the collected sample means
# According to the Central Limit Theorem, this distribution should approximate a normal distribution
plt.hist(sample_means, bins=30)
plt.title("Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

## Summary of the Worksheet

This notebook effectively illustrates two cornerstone theorems of statistics:

### 1. The Law of Large Numbers (LLN)

*   **Demonstration:** Through simulations of coin tosses, dice rolls, and sampling from random integers (1-9), the notebook clearly shows that as the number of trials or sample size increases, the *running mean* of the observations converges towards the *theoretical expected value* of the underlying distribution.
*   **Key Takeaway:** The more data we collect, the closer our observed average will be to the true average. This is evident in the plots where running means stabilize around the expected values (0.5 for coin, 3.5 for dice, and 5.0 for 1-9 integers) as the number of samples grows.

### 2. The Central Limit Theorem (CLT)

*   **Demonstration:** Starting with a simple uniform distribution (non-normal), the notebook then takes multiple samples from this distribution, calculates the mean of each sample, and plots the distribution of these sample means.
*   **Key Takeaway:** Regardless of the shape of the original population distribution (as long as it has a finite mean and variance), the distribution of sample means will tend towards a *normal distribution* as the sample size increases. This is a powerful concept, as it allows us to use normal distribution-based statistical methods even when dealing with non-normal data, provided our sample sizes are sufficiently large.
